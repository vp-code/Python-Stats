```python
import numpy as np
import pandas as pd
import scipy.stats as ss
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x = ss.norm.rvs(size=1000,loc=1,scale=1)
y = ss.norm.rvs(size=1000,loc=4,scale=1)

ax = sns.kdeplot(x,shade=True)
ax = sns.kdeplot(y,shade=True)

ax.tick_params(bottom=False,left=False,labelbottom=False,labelleft=False)

_=ax.update({'xlabel':'Income','title':'Hypothetical income distributions'})
```
