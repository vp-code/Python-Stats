```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
iris = sns.load_dataset("iris")
iris.info()
```

```python
# Save the DataFrame

iris.to_pickle("/home/vpoduri/python/iris.pkl")
```

```python
# Simple scatter plot

_ = sns.scatterplot(data=iris,x="sepal_width",y="sepal_length",hue="species")
```

```python
# Line plot
# If there are multiple y values for given x, the default plot contains a 95% confidence interval band

_ = sns.lineplot(data=iris,x="sepal_width",y="sepal_length",hue="species")

# Better to plot sd (pandas DataFrame method called)

_ = sns.lineplot(data=iris,x="sepal_width",y="sepal_length",hue="species",ci="sd")
```

```python
# Box plot
_ = sns.boxplot(data=iris,x="species",y="petal_length")
```

```python
# Univariate Distribution
_ = sns.distplot(iris.petal_length)
```

```python
# Bivariate distribution
_ = sns.jointplot(data=iris,x="petal_length",y="sepal_length")
```

```python
# Multivariate distributions
_ = sns.pairplot(iris,hue="species")

# You can limit the variables plotted

_ = sns.pairplot(iris,vars=["petal_width","petal_length"],hue="species")
```
