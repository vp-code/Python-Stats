```python
fig = plt.figure()           # Initiate figure
a1 = fig.add_subplot(221)    # Adding axes. Argument means figure with 2 rows, two columns; select subplot 1

n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

a1.scatter(X,Y)
a1.axhline(y=1,color='red')         # Add horizontal line. 
                                    # Colors from https://matplotlib.org/examples/color/named_colors.html

x =  np.linspace(-2*np.pi, 2*np.pi, 1000)

a2 = fig.add_subplot(224)    # Figure has 2 rows, 2 columns; select subplot 4
a2.plot(x,np.sin(x))
a2.grid()
plt.show()
```
